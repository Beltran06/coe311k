```python
import random as rand
import math
import matplotlib.pyplot as plt

def monte_carlo_pi(number_of_shots):
    number_of_hits = 0
    for i in range (0, number_of_shots):
        x = rand.random()
        y = rand.random()
        
        if (x*x + y*y < 1):
            number_of_hits = number_of_hits +1
            
    return 4*number_of_hits/number_of_shots

monte_carlo_pi(1000)
```

```python
for i in range(0,100):
    calculated_pi = monte_carlo_pi(1000)
    true_pi = math.pi
    true_error = abs(true_pi - calculated_pi)
    relative_error = abs(true_pi - calculated_pi)/(true_pi)
    
    print(calculated_pi, true_error, relative_error)
```
