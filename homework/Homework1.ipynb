```python
import random as rand
import math
import matplotlib.pyplot as plt

def monte_carlo_pi(number_of_shots):
    number_of_hits = 0
    for i in range (0, number_of_shots):
        x = rand.random()
        y = rand.random()
        
        if (x*x + y*y < 1):
            number_of_hits = number_of_hits +1
            
    return 4*number_of_hits/number_of_shots

monte_carlo_pi(1000)
```

```python
true_pi = math.pi

sample_sizes = []
absolute_errors = []

for i in range(10,1001, 20):
    calculated_pi = monte_carlo_pi(i)
    error = abs(calculated_pi - true_pi)
    
    sample_sizes.append(i)
    absolute_errors.append(error)
```

```python
plt.figure()
plt.plot(sample_sizes, absolute_errors, 'o-')
plt.xlabel("Number of Samples")
plt.ylabel("Absolute Error")
plt.title("Monte Carlo Pi Absolute Error vs Number of Samples")
plt.show()
```
