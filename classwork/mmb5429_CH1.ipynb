```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 50)

plt.subplot(2, 2, 1) # (row, column, active area)
plt.plot(x, np.sin(x), 'r')
plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x), 'g')
plt.subplot(2, 2, 3)
plt.plot(x, np.sin(x), 'b')
plt.subplot(2, 2, 4)
plt.plot(x, np.cos(x), 'y')
plt.show()
```
