```python
import numpy as np

N = 5000

pi_approx = 0

for n in range(N):
    term = (-1)**n / (2*n + 1)
    pi_approx += term
    
    current_pi = 4*pi_approx
    
    if (n+1)%100 == 0:
        relative_error = abs(current_pi - np.pi)/np.pi
        print(f"Term:{n+1:4d}, pi = {current_pi:.10f}, Relative Error = {relative_error:.6e}")
```
