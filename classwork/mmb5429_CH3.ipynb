```python
import random as rand
import math
import matplotlib.pyplot as plt

def monte_carlo_pi(number_of_shots):
    number_of_hits = 0
    for i in range (0, number_of_shots):
        x = rand.random()
        y = rand.random()
        
        if (x*x + y*y < 1):
            number_of_hits = number_of_hits +1
            
    return 4*number_of_hits/number_of_shots

monte_carlo_pi(1000)
```

```python
absolute_errors = []
relative_errors = []

for i in range(0,100):
    calculated_pi = monte_carlo_pi(1000)
    true_pi = math.pi
    true_error = abs(true_pi - calculated_pi)
    relative_error = abs(true_pi - calculated_pi)/(true_pi)
    
    absolute_errors.append(true_error)
    relative_errors.append(relative_error)
```

```python
plt.figure()
plt.plot(absolute_errors, 'o')
plt.xlabel("Run Number")
plt.ylabel("Absolute Error")
plt.title("Absolute Error vs Run Number")
plt.show()

plt.figure()
plt.plot(relative_errors, 'o')
plt.xlabel("Run Number")
plt.ylabel("Relative Error")
plt.title("Relative Error vs Run Number")
plt.show()
```
